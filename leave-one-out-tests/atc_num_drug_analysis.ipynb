```python
import pandas as pd
atc_numbers = pd.read_csv("./../data/full database/drugbank_drug_and_first_letter_atc_codes.csv")
print(atc_numbers["First letter ATC Codes"].value_counts(normalize=True))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
atc_numbers["First letter ATC Codes"].value_counts().plot(kind='bar')
plt.xlabel('EC number')
plt.ylabel('Frequency')
plt.title('Distribution of EC numbers in known drug targets')
plt.show()
```
